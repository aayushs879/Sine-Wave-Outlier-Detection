```python
#this is the notebook for analysing the sine wave and recognizing the sample which is not on the wave
#please note that sample no.s are used python indices therefore 1 idx means 2nd sample 

#importing libraries
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline



#defining a function to get a sine wave
def get_wave(phase, frequency):
    n_samples = 100 #since each time we will have 100 samples
    fc = frequency
    fs = fc*n_samples
    x = np.arange(fs)
    
    y = [np.sin(2*np.pi*fc*(i/fs) + phase) for i in x]
    
    plt.stem(x[:100], y[:100])#plotting the wave for one cycle
    plt.show()
    
    return x[:100], y[:100] #returning only first cycle
    
```

After a lot of experimentation on this exercise it turned out that no matter whatever the frequency is, if we always want 100 samples of a sine wave, then with a constant phase, the values of samples will remain same for any frequency.

```python
#altering the sample so that it doesn't lies on the wave
def alter_sample(y, idx, factor, reduce):
    if reduce == True:
        y[idx] = y[idx]*(1 - factor)
    else:
        y[idx] = y[idx]*(1 + factor)
    
    plt.stem(np.arange(len(y)), y)
    plt.show()
    return y 
    
```

```python
def odd_sample(x, y):
    for i in range(len(y)):
        if not (-1<= y[i] <= 1):
            X = i
            return X 
    
    #calculating inverse sine for each sample
    p = []
    for i in range(len(y)):
        p.append(math.asin(y[i]) - 2*np.pi*(i+1)/100)
    
    #calculating slope of inverse values between two samples
    l = []
    for i in range(len(p)-1):
        l.append(round(p[i+1] - p[i], 4))
    
    sns.lineplot(np.arange(len(l)), l, )#plotting the graph
    
    # outlier in the graph is easily visible with that abrupt change 
    # but still we have to take the value of sample which corresponds to that
    # so we take a derivative of this graph and at the place where we two consecutive sign change
    # that place will correspond to that no. of sample
    
    
    #the derivative of the previous graph
    m = []
    for i in range(len(l)- 1):
        m.append(l[i+1]- l[i])
        if(m[i] == 0.0):
            m[i] = 0
            
            
    #special case for first and 98th value
    if ((m[0]<0) and (m[1]>0) and m[2] ==0) or ((m[0]>0) and (m[1]<0) and m[2] ==0):
        return 1
    
    if ((m[-1]<0) and (m[-2]>0) and m[-3] ==0) or ((m[-1]>0) and (m[-2]<0) and m[-3] ==0):
        return 98
    
    
    #special case for zeroth and 99th value
    if ((m[0]<0) and (m[1]==0)) or ((m[0]>0) and m[2] ==0):
        return 0
    if ((m[-1]<0) and (m[-2]==0)) or ((m[-1]>0) and m[-2] ==0):
        return 99
    
    # for all other cases trying to check for sign changes
    for i in range(len(m)-2):
        if((m[i]<0) and (m[i+1]>0) and (m[i+2]<0)) or ((m[i]>0) and (m[i+1]<0) and (m[i+2]>0)):
            oddval = i + 2 # here we have +2 as previously using two derivatives the index's value had come earlier
            return oddval
    
    
```

```python
x, y = get_wave(np.pi/87, 33)
y = alter_sample(y, 1, 0.20, False)
p = odd_sample(x, y)
print('The outlying sample was = ', p)
```
